```python
from datetime import date

#2
df = pd.read_csv('wells_info.csv')
df[['PermitDate','CompletionDate']] = df[['PermitDate','CompletionDate']].apply(pd.to_datetime)
df['work duration'] = (df['CompletionDate'] - df['PermitDate']) / np.timedelta64(1, 'M')
df['work duration'] = df['work duration'].apply(np.floor)
display(df.head())
```

```python
#3
df = pd.read_csv('wells_info_na.csv')
df['CompletionDate'] = pd.to_datetime(df['CompletionDate'])
df['FirstProductionDate'] = pd.to_datetime(df['FirstProductionDate'])

freq = {}
for column in df:
    freq.update(dict(df[column].value_counts()))
freq = dict(sorted(freq.items(), key=lambda item: item[1]))
most_common = list(freq.keys())[-1]

df = df.fillna(df.median(numeric_only=True))
df['CompletionDate'] = df['CompletionDate'].median()
df = df.fillna(most_common)
display(df)
```

```python
#4
df1 = pd.read_csv('wells_info.csv')
df2 = pd.read_csv('production.csv')

a = df2.groupby('API').sum()
extract  = a['Liquid']
df1 = df1.join(extract, on = 'API', how='left', lsuffix='_left', rsuffix='_right')
display(df1.head())
```
