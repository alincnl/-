```python
import pandas as pd
import numpy as np
import datetime
from datetime import time

#4.1

df1 = pd.read_csv('cinema_sessions.csv',sep= ' ', index_col=0)
df2 = pd.read_csv('titanic_with_labels.csv', sep = '\s+',index_col=0)
display(df2.head(100))
```

```python
    #1 пусть 0 - муж, 1 - жен
print("sex filter")
df2 = df2[(df2['sex'] == 'ж') | (df2['sex'] == 'Ж')|(df2['sex'] == 'м')|(df2['sex'] == 'М')]
df2['sex']=df2['sex'].map(lambda value: 1 if value=='ж' or value=='Ж' else 0)
display(df2.head(100))
```

```python
    #2
df2['row_number']=df2['row_number'].fillna(value=max(df2['row_number']))
display(df2.head(100))
```

```python
    #3 пусть нереально большое значение - 10
avg = df2['liters_drunk'].mean()
df2['liters_drunk']=df2['liters_drunk'].map(lambda value: avg if value<0 or value>10 else value)
display(df2.head(100))
```

```python
#4.2

    #4
df2["age_<18"] = np.where(df2['age']< '18', df2['age'], False)
df2["age_18-50"] = np.where((df2['age']>= '18') & (df2['age']<= '50'), df2['age'], False)
df2["age_>50"] = np.where(df2['age']> '50', df2['age'], False)
df2 = df2.drop(['age'], axis=1)
display(df2.head(100))
```

```python
    #5 0 - no, 1 - yes
print("WITHOUT ALCO")
df2['drink']=np.where(df2['drink'].astype(str).str.contains('beer'), 1, 0)
display(df2.head(100))
```

```python
    #6
df3 = pd.merge(df1, df2, on='check_number')
display(df3.head(100))
```

```python
df3["morning"] = np.where(pd.to_datetime(df3['session_start'])< datetime.datetime.now().replace(hour=12, minute=0, second=0, microsecond=0), df3['session_start'], False)
df3["day"] = np.where((pd.to_datetime(df3['session_start'])> datetime.datetime.now().replace(hour=12, minute=0, second=0, microsecond=0)) & (pd.to_datetime(df3['session_start'])< datetime.datetime.now().replace(hour=18, minute=0, second=0, microsecond=0)), df3['session_start'], False)
df3["evening"] = np.where(pd.to_datetime(df3['session_start'])>= datetime.datetime.now().replace(hour=18, minute=0, second=0, microsecond=0), df3['session_start'], False)
display(df3.head(100))
```
